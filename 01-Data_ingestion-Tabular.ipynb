```python
# Doign a conversion of letter grades to numbers 
```

```python
def num_score(x):
    to_num = {
        "A+" : 4.3, "A" : 4.0, "A-" : 3.7,
        "B+" : 3.3, "B" : 3.0, "B-" : 2.7,
        "C+" : 2.3, "C" : 2.0, "C-" : 1.7
    }
    return x.map(lambda x : to_num.get(x,x))
```

```python
(students
.set_index('Last Name')
.drop('First Name', axis=1)
.apply(num_score)
.T
.plot(title="Student score by year")
.legend(bbox_to_anchor=(1, .75))
);
```

```python
students.melt(
id_vars=["Last Name", "First Name"],
var_name="Level",
value_name="Score"
).set_index(['First Name', 'Last Name', 'Level'])
```
